```python
def aug(shape):
    a = tf.truncated_normal(shape, stddev = 0.1)
    return tf.Variable(a)
def bia(shape):
    a = tf.constant(0.1, shape = shape)
    return tf.Variable(a)
def conv2d(x, w):
    return tf.nn.conv2d(x, w, strides = [1, 1, 1, 1], padding = 'SAME')
def max_pool(x):
    return tf.nn.max_pool(x, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1],\
                         padding = 'SAME')
```

```python
def batches(input_=None, label_=None, batch_size=None, shuffle=False):
    assert len(input_) == len(label_)
    if shuffle:
        ind = np.arange(len(input_))
        np.random.shuffle(ind)
    for i in range(0, len(input_) - batch_size + 1, batch_size):
        if shuffle:
            e = ind[i:i + batch_size]
        else:
            e = slice(i, i + batch_size)
        yield input_[e], label_[e]
```

```python
sess = tf.InteractiveSession()
w1 = aug([2, 2, 1, 16])
b1 = bia([16])
input_ = tf.placeholder(tf.float32, [None, 8])
label_ = tf.placeholder(tf.float32, [None, 1])
keep_prob = tf.placeholder(tf.float32)
xi = tf.reshape(input_, [-1, 4, 2, 1])
x_shortcut = xi
h1 = tf.nn.relu(conv2d(xi, w1) + b1)

w2 = aug([2, 2, 16, 32])
b2 = bia([32])
h2 = tf.nn.relu(conv2d(h1, w2) + b2)

w_shortcut = aug([1,1, 1, 4*2*4])
x_shortcut = tf.nn.conv2d(x_shortcut, w_shortcut,\
                strides=[1, 1, 1, 1], padding='SAME')

add = tf.add(x_shortcut, h2)
prediction_add = tf.nn.relu(add)


w_fc1 = aug([4*2*32, 512])
b_fc1 = bia([512])

h_pfc1 = tf.reshape(prediction_add, [-1, 4*2*32])
h_fc1 = tf.nn.relu(tf.matmul(h_pfc1, w_fc1)+b_fc1)
h_dfc1 = tf.nn.dropout(h_fc1, keep_prob)



w_fc2 = aug([512, 1])
b_fc2 = bia([1])
prediction = tf.matmul(h_dfc1, w_fc2) + b_fc2




cross_entropy = tf.reduce_mean(tf.reduce_sum(tf.square(label_ - prediction), \
                                reduction_indices = [1]))
correct_prediction = tf.equal(prediction, label_)
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float64))
train_step = tf.train.AdamOptimizer(0.01).minimize(cross_entropy)


sess = tf.Session()
sess.run(tf.global_variables_initializer())
n_epoch = 30
batch_size = 5
#     sess.run(train_step, feed_dict ={input_:x_train, label_:y_train, \
#                                      keep_prob: 1.0})
#     print(i, 'loss = ', sess.run(cross_entropy, \
#                     feed_dict ={input_:x_train, label_:y_train, \
#                                      keep_prob: 1.0}))   
prediction_value = sess.run(prediction, feed_dict = {input_:x_test, \
                                    label_:y_test, \
                                      keep_prob: 1.0})
    # training
for e in range(n_epoch):
    train_loss, train_acc, n_batch = 0, 0, 0
    for i, j in batches(x_train1, y_train1, batch_size, shuffle=True):
        _ = tf.placeholder(tf.float32)
        _, err, ac = sess.run([train_step, cross_entropy, accuracy], feed_dict={input_: i, label_: j, keep_prob: 1.0})
        train_loss += err
        train_acc += ac
        n_batch += 1
    print("train loss: %f" % (train_loss / n_batch))
    print("train accuracy: %f" % (train_acc / n_batch))

    # validation
#     val_loss, val_acc, n_batch = 0, 0, 0
#     for i, j in batches(x_test, y_test, batch_size, shuffle=False):
#         err, ac = sess.run([cross_entropy, accuracy], feed_dict={input_: i, label_: j, keep_prob: 1.0})
#         val_loss += err
#         val_acc += ac
#         n_batch += 1
#     print("validation loss: %f" % (val_loss / n_batch))
#     print("validation accuracy: %f" % (val_acc / n_batch))
    #print('*' * 50)
sess.close()
```
